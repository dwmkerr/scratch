# Sales Analysis

This analysis loads our sales data, then shows it as a table. It then calculates the sum of the sales. Finally, it visualises the sales as a line chart with a trendline.

```python
import pandas as pd

# Load our sales CSV data into a dataframe then print it.
df = pd.read_csv('ch01-sales.csv')
print(df)

# The total sales is the sum of the 'amount' column. Calculate and print it.
total_sales = df['Amount (USD)'].sum()
print(f'Total Sales: ${total_sales}')

import matplotlib.pyplot as plt

# Create the line chart
plt.figure(figsize=(5, 3))
plt.plot(df['Date'], df['Amount (USD)'])
plt.xlabel('Date')
plt.ylabel('Amount (USD)')
plt.title('Sales Amount Over Time')
plt.show()
```
